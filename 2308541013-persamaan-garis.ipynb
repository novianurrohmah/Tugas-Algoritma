```python
import numpy as np 
import matplotlib.pyplot as plt 

#Persamaan garis pertama melalui titik (p,q) dan (r,s)

# 1.Masukkan nilai p,q,r dan s
p_13 = 3 #Titik(p,q)
q_13 = 0
r_13 = 0 #Titik(r,s)
s_13 = 3

# 2.Menghitung gradien persamaan garis pertama
m1_13=(s_13-q_13)/(r_13-p_13)


# 3.Memunculkan hasil
print("Gradien dari persamaan garis pertama adalah {}".format(m1_13))
print("Persamaan garis pertama adalah y={}x+{}".format(m1_13,-m1_13*p_13+q_13))
#y1_13=m1_13*(x-p_13)+q_13

x=np.arange(-8,8,0.01)
plt.plot(x,m1_13*(x-p_13)+q_13,color='yellow')
plt.show()
```

```python
#Persamaan garis kedua melalui titik (w,x) dan (y,z)

# 1.Masukkan nilai w,x,y, dan z
w_13 = 9 #Titik (w,x)
x_13 = 0
y_13 = 0 #Titik (y,z)
z_13 = 9

# 2.Menghitung gradien persamaaan garis kedua
m2_13 =(z_13-x_13)/(y_13-w_13)

# 3.Memunculkan hasil
print("Gradien dari persamaan garis kedua adalah {}".format(m2_13))
print("Persamaan garis kedua adalah y={}x+{}".format(m2_13,-m2_13*w_13+x_13))
#y2_13=m2_13*(x-w_13)+x_13

x=np.arange(-8,8,0.01)
plt.plot(x,m2_13*(x-w_13)+x_13,color='red')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

# Persamaan garis yang memalui 2 titik yaitu titik (x1,y1) dan (x2,y2)
# Input titik pertama (x1,y1)
x1_13 = float(input("Masukkan nilai x1: "))
y1_13 = float(input("Masukkan nilai y1: "))

# Input titik kedua
x2_13 = float(input("Masukkan nilai x2: "))
y2_13 = float(input("Masukkan nilai y2: "))

# Menghitung gradien
m_13 = (y2_13 - y1_13)/(x2_13 - x1_13)

print("Gradien dari persamaan garis tersebut adalah {}".format(m_13))
print("Persamaan garis tersebut adalah y={}x+{}".format(m_13,-m_13*x1_13+y1_13))
#y_13=m_13*(x-x1_13)+y1_13

x=np.arange(-8,8,0.01)
plt.plot(x,m_13*(x-x1_13)+y1_13,color='black')
plt.show()
```

```python
#mPersamaan garis pertama yang melalui titik (x1,y1) dan (x2,y2)
# Input titik-titik garis pertama
x1_13 = float(input("Masukkan nilai x1 untuk garis pertama: "))
y1_13 = float(input("Masukkan nilai y1 untuk garis pertama: "))
x2_13 = float(input("Masukkan nilai x2 untuk garis pertama: "))
y2_13 = float(input("Masukkan nilai y2 untuk garis pertama: "))

# Mencari gradien garis pertama
m1_13 = (y2_13 - y1_13) / (x2_13 - x1_13)

print("Gradien garis pertama adalah {}".format(m1_13))
print("Persamaan garis pertama adalah y={}x+{} ".format(m1_13,-m1_13*x1_13+y1_13))
#y = m1_13*(x-x1_13)+y1_13


#Persamaan garis kedua yang melalui titik (a1,b1) dan (a2,b2)
# Input titik-titik garis kedua
a1_13 = float(input("Masukkan nilai a1 untuk garis kedua: "))
b1_13 = float(input("Masukkan nilai b1 untuk garis kedua: "))
a2_13 = float(input("Masukkan nilai a2 untuk garis kedua: "))
b2_13 = float(input("Masukkan nilai b2 untuk garis kedua: "))

# Mencari gradien garis kedua
m2_13 = (b2_13 - b1_13) / (a2_13 - a1_13)

print("Gradien garis kedua adalah {}".format(m2_13))
print("Persamaan garis kedua adalah y={}x+{} ".format(m2_13,-m2_13*a1_13+b1_13))
#y = m2_13*(x-a1_13)+b1_13

#Mencari titik potong
if m1_13 == m2_13:
    print("Wahh ternyata gradiennya sama, jadi kemungkinan gaada solusinya atau bisa juga solusinya banyak.") 
else:
    # Titik potong kedua garis pada (tp_x,tp_y)
    tp_x = (m1_13*x1_13-y1_13-m2_13*a1_13+b1_13)/(m1_13-m2_13)
    tp_y = m2_13*(tp_x-a1_13)+b1_13
    
    print ("titik potong nya adalah ({},{})".format(tp_x,tp_y))
    x=np.arange(tp_x-7,tp_x+7,0.01)
    plt.figure(figsize=(4,4),dpi=500)
    
    plt.plot(x,m1_13*(x-x1_13)+y1_13,color='orange',label='y={}x+{}'.format(m1_13,-m1_13*x1_13+y1_13))
    plt.plot(x,m2_13*(x-a1_13)+b1_13,color='red',label='y={}x+{}'.format(m2_13,-m2_13*a1_13+b1_13))
    plt.plot(tp_x,tp_y,color='black', marker='o',markersize=5)
    plt.legend(loc='best')
    plt.show()
```
