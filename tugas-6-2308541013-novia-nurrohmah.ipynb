```python
matriks_13 = np.array([[10,4,8,3],
                       [2,6,1,9],
                       [4,0,1,5]])

nilai_min_13 = np.min(matriks_13)
min_kolom_13 = np.min(matriks_13, axis=0)
min_baris_13 = np.min(matriks_13, axis=1)


nilai_max_13 = np.max(matriks_13)
max_kolom_13 = np.max(matriks_13, axis=0)
max_baris_13 = np.max(matriks_13, axis=1)

nilai_sum_13 = np.sum(matriks_13)
sum_kolom_13 = np.sum(matriks_13, axis=0)
sum_baris_13 = np.sum(matriks_13, axis=1)

print("Nilai minimum:", nilai_min_13)
print("Nilai minimum per kolom:", min_kolom_13)
print("Nilai minimum per baris:", min_baris_13)

print("Nilai maksimum:", nilai_max_13)
print("Nilai maksimum per kolom:", max_kolom_13)
print("Nilai maksimum per baris:", max_baris_13)

print("Jumlah:", nilai_sum_13)
print("Jumlah per kolom:", sum_kolom_13)
print("Jumlah per baris:", sum_baris_13)
```
