```python
kecepatan_motor_1 = float(input("Masukkan kecepatan motor pertama (dalam km/jam) = "))

kecepatan_motor_2 = float(input("Masukkan kecepatan motor kedua (dalam km/jam) = "))

if kecepatan_motor_1 > kecepatan_motor_2 :
    print("Motor pertama dengan kecepatan",kecepatan_motor_1,"km/jam lebih cepat dari motor kedua dengan kecepatan",kecepatan_motor_2,"km/jam.")
elif kecepatan_motor_1 < kecepatan_motor_2 :
    print("Motor pertama dengan kecepatan",kecepatan_motor_1,"km/jam lebih lambat dari motor kedua dengan kecepatan",kecepatan_motor_2,"km/jam.")
else:
    print("Motor pertama dengan kecepatan",kecepatan_motor_1,"km/jam sama dengan motor kedua dengan kecepatan",kecepatan_motor_2,"km/jam.")
```

```python
suhu_dalam_celcius = float(input("Masukkan suhu dalam Celcius = "))
suhu_dalam_fahrenheit = (suhu_dalam_celcius * 9/5) + 32
suhu_dalam_reaumur = suhu_dalam_celcius * 4/5
suhu_dalam_kelvin = suhu_dalam_celcius + 273.15

print("Suhu dalam Celcius =", suhu_dalam_celcius)
print("Suhu dalam Fahrenheit =", suhu_dalam_fahrenheit)
print("Suhu dalam Reaumur =", suhu_dalam_reaumur)
print("Suhu dalam Kelvin =", suhu_dalam_kelvin)
```

```python
jarak = float(input("Masukkan jarak dalam kilometer = "))
waktu = float(input("Masukkan waktu dalam jam = "))
kecepatan = jarak / waktu
print("Kecepatan rata-rata adalah", kecepatan, "kilometer/jam.")
```

```python
harga_barang = float(input("Masukkan harga barang = "))
diskon = float(input("Masukkan persentase diskon (dalam %)= "))
jumlah_diskon = (diskon / 100) * harga_barang
harga_setelah_diskon = harga_barang - jumlah_diskon

print("Jumlah diskon adalah",jumlah_diskon,"Rupiah")
print("Harga setelah diskon adalah",harga_setelah_diskon,"Rupiah")
```

```python
alas = float(input("Masukkan nilai alas segitiga ="))
tinggi = float(input("Masukkan nilai tinggi segitiga ="))

luas = 0.5 * alas * tinggi

print("Luas segitiga = ",luas)
```

```python
import numpy as np
phi = np.pi
jari_jari = 4.5 
luas = phi * jari_jari ** 2
keliling = 2 * phi * jari_jari
print("Luas Lingkaran =",luas)
print("Keliling lingkaran =",keliling)
```
