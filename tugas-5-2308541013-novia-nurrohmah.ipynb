```python
import numpy as np

def hitung_luas_segitiga_13(alas_13, tinggi_13):
    luas_13 = 0.5 * alas_13 * tinggi_13
    return luas_13

def hitung_luas_lingkaran_13(jari_jari_13):
    luas_13 =np.pi * jari_jari_13 ** 2
    return luas_13

def hitung_luas_persegi_13(sisi_13):
    luas_13 = sisi_13 ** 2
    return luas_13

def hitung_luas_persegi_panjang_13(panjang_13, lebar_13):
    luas_13 = panjang_13 * lebar_13
    return luas_13

def hitung_luas_bidang_13(bidang_13):
    if bidang_13.upper() == "SEGITIGA":
        alas_13 = float(input("Masukkan panjang alas segitiga: "))
        tinggi_13 = float(input("Masukkan tinggi segitiga: "))
        luas_13 = hitung_luas_segitiga_13(alas_13, tinggi_13)
        print(f"Luas segitiga adalah: {luas_13}")
    elif bidang_13.upper() == "LINGKARAN":
        jari_jari_13 = float(input("Masukkan jari-jari lingkaran: "))
        luas_13 = hitung_luas_lingkaran_13(jari_jari_13)
        print(f"Luas lingkaran adalah: {luas_13}")
    elif bidang_13.upper() == "PERSEGI":
        sisi_13 = float(input("Masukkan panjang sisi persegi: "))
        luas_13 = hitung_luas_persegi_13(sisi_13)
        print(f"Luas persegi adalah: {luas_13}")
    elif bidang_13.upper() == "PERSEGI PANJANG":
        panjang_13 = float(input("Masukkan panjang persegi panjang: "))
        lebar_13 = float(input("Masukkan lebar persegi panjang: "))
        luas_13 = hitung_luas_persegi_panjang_13(panjang_13, lebar_13)
        print(f"Luas persegi panjang adalah: {luas_13}")
    else:
        print("Maaf, bidang tersebut tidak termasuk dalam perhitungan.")

bidang_13 = input("Masukkan jenis bidang (segitiga, lingkaran, persegi, atau persegi panjang): ")
hitung_luas_bidang_13(bidang_13)
```
