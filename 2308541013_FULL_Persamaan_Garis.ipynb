```python
import numpy as np
import matplotlib.pyplot as plt

def hitung_persamaan_garis_dua_titik(x1_13, y1_13, x2_13, y2_13):
    m_13 = (y2_13 - y1_13) / (x2_13 - x1_13)
    c_13 = y1_13 - m_13 * x1_13
    return m_13, c_13

def hitung_persamaan_garis_gradien_dan_titik(m_13, x_13, y_13):
    c_13 = y_13 - m_13 * x_13
    return m_13, c_13

def hitung_persamaan_garis_persamaan(a_13, b_13, c_13):
    m_13 = -a_13 / b_13
    c_13 = -c_13 / b_13
    return m_13, c_13

def hitung_titik_potong(m1_13, c1_13, m2_13, c2_13):
    if m1_13 == m2_13:
        return None

    x_13 = (c2_13 - c1_13) / (m1_13 - m2_13)
    y_13 = m1_13 * x_13 + c1_13
    return x_13, y_13

def plot_grafik_garis(x_13, y_13, label):
    plt.plot(x_13, y_13, label=label)

def main():
    kondisi1 = input("Masukkan kondisi yang diketahui untuk garis pertama (titik/gradien/persamaan): ")

    if kondisi1 == "titik":
        x1_13 = float(input("Masukkan nilai x dari titik pertama: "))
        y1_13 = float(input("Masukkan nilai y dari titik pertama: "))
        x2_13 = float(input("Masukkan nilai x dari titik kedua: "))
        y2_13 = float(input("Masukkan nilai y dari titik kedua: "))

        m1_13, c1_13 = hitung_persamaan_garis_dua_titik(x1_13, y1_13, x2_13, y2_13)
    elif kondisi1 == "gradien":
        m1_13 = float(input("Masukkan nilai gradien: "))
        x_13 = float(input("Masukkan nilai x dari titik yang dilewati: "))
        y_13 = float(input("Masukkan nilai y dari titik yang dilewati: "))

        m1_13, c1_13 = hitung_persamaan_garis_gradien_dan_titik(m1_13, x_13, y_13)
    elif kondisi1 == "persamaan":
        a_13 = float(input("Masukkan nilai a: "))
        b_13 = float(input("Masukkan nilai b: "))
        c_13 = float(input("Masukkan nilai c: "))

        m1_13, c1_13 = hitung_persamaan_garis_persamaan(a_13, b_13, c_13)
    else:
        print("Kondisi yang diketahui untuk garis pertama tidak valid.")
        return

    kondisi2 = input("Masukkan kondisi yang diketahui untuk garis kedua (titik/gradien/persamaan): ")

    if kondisi2 == "titik":
        x1_13 = float(input("Masukkan nilai x dari titik pertama: "))
        y1_13 = float(input("Masukkan nilai y dari titik pertama: "))
        x2_13 = float(input("Masukkan nilai x dari titik kedua: "))
        y2_13 = float(input("Masukkan nilai y dari titik kedua: "))

        m2_13, c2_13 = hitung_persamaan_garis_dua_titik(x1_13, y1_13, x2_13, y2_13)
    elif kondisi2 == "gradien":
        m2_13 = float(input("Masukkan nilai gradien: "))
        x_13 = float(input("Masukkan nilai x dari titik yang dilewati: "))
        y_13 = float(input("Masukkan nilai y dari titik yang dilewati: "))

        m2_13, c2_13 = hitung_persamaan_garis_gradien_dan_titik(m2_13, x_13, y_13)
    elif kondisi2 == "persamaan":
        a_13 = float(input("Masukkan nilai a: "))
        b_13 = float(input("Masukkan nilai b: "))
        c_13 = float(input("Masukkan nilai c: "))

        m2_13, c2_13 = hitung_persamaan_garis_persamaan(a_13, b_13, c_13)
    else:
        print("Kondisi yang diketahui untuk garis kedua tidak valid.")
        return

    print("Persamaan garis pertama adalah y = {}x + {}".format(m1_13, c1_13))
    print("Persamaan garis kedua adalah y = {}x + {}".format(m2_13, c2_13))

    titik_potong_13 = hitung_titik_potong(m1_13, c1_13, m2_13, c2_13)
    if titik_potong_13 is None:
        print("Kedua garis ini parallel dan tidak memiliki titik potong.")
    else:
        print("Titik potong kedua garis adalah ({:.2f}, {:.2f})".format(titik_potong[0], titik_potong[1]))

    x_13 = np.linspace(-10, 10, 100)
    y1_13 = m1_13 * x_13 + c1_13
    y2_13 = m2_13 * x_13 + c2_13

    plot_grafik_garis(x_13, y1_13, "Garis Pertama")
    plot_grafik_garis(x_13, y2_13, "Garis Kedua")

    if titik_potong_13 is not None:
        plt.plot(titik_potong[0], titik_potong[1], 'ro', label="Titik Potong")

    plt.xlabel("Nilai x")
    plt.ylabel("Nilai y")
    plt.title("Grafik Dua Garis")
    plt.legend()
    plt.grid(True)
    plt.show()

main()
```
