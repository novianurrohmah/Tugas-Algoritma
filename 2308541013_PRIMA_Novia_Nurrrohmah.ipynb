```python
#Soal
print("Jika diketahui bahwa ğ‘› adalah dua digit terakhir dari NIM kalian, maka tentukan biangan prima antara 1000ğ‘› dan 1000(ğ‘›+1).")

#Mengecek bilangan prima
def cek_bil_prima_13(angka):
    if angka_13 <= 1:
        return False
    elif angka_13 <= 3:
        return True
    elif angka_13 % 2 == 0 or angka_13 % 3 == 0:
        return False
    else:
        i_13 = 5
        while i_13 * i_13 <= angka_13:
            if angka_13 % i_13 == 0 or angka_13 % (i_13 + 2) == 0:
                return False
            i_13 += 6
        return True

#Diketahui n adalah 2 digit terakhir nim dan dicari batas nya
n_13 = 13
batas_bawah_13 = 1000 * n_13
batas_atas_13 = 1000 * (n_13 + 1)

#Meunculkan bilangan prima diantara rentan yang sudah ditentukan
bil_prima_13 = []
for angka_13 in range(batas_bawah_13, batas_atas_13):
    if cek_bil_prima_13(angka_13):
        bil_prima_13.append(angka_13)

#Menghitung jumlah bilangan prima
jumlah_bil_prima_13 = 0
for angka_13 in range(batas_bawah_13, batas_atas_13):
    if cek_bil_prima_13(angka_13):
        jumlah_bil_prima_13 += 1

#Memunculkan hasil
print("Karena n adalah 2 digit terakhir nim, maka n =", n_13, ".Sehingga diperoleh batas bawahnya adalah", batas_bawah_13, "dan batas atasnya adalah ", batas_atas_13, "\n")
print("Maka")
print("Bilangan prima antara", batas_bawah_13, "dan", batas_atas_13, "adalah:")
print(bil_prima_13, "\n")
print("Jumlah bilangan prima antara", batas_bawah_13, "dan", batas_atas_13,"adalah sebanyak", jumlah_bil_prima_13)
```
